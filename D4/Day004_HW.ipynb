### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
import json
import requests

r = requests.get('https://cat-fact.herokuapp.com/facts')
print(r.text)
print(json.loads(r.text))
```
